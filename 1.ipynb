```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area ')
plt.ylabel('price ')
plt.scatter(df.area, df.price, color='black', marker='^')
```

```python
df1= df.drop('price', axis='columns')
print(df1)

price=df.price
price
```

```python
reg= linear_model.LinearRegression()
reg.fit(df1, price)
```

```python
reg.predict([[3200]])
```

```python
print(reg.coef_)
print(reg.intercept_)
print(135.78767123*3200 + 180616.43835616432)
```

```python
dfx= pd.read_csv("areas.csv")
print(dfx)
print(dfx.head(4))
```

```python
p=reg.predict(dfx)
p
```

```python

dfx['prices']=p
print(dfx)
```
